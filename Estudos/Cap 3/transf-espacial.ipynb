```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from matplotlib.colors import NoNorm
from skimage import io
```

```python
# Load an image
img = cv2.imread('normal-mammography.png')

# Define the lower and upper limits of the input range
lower_limit = 50
upper_limit = 200

# Define the lower and upper limits of the output range
out_lower_limit = 0
out_upper_limit = 255

# Normalize the image using cv2.normalize function
img_norm = cv2.normalize(img, None, alpha=out_lower_limit, beta=out_upper_limit, 
                         norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)

# Apply contrast stretching
img_adj = cv2.normalize(img_norm, None, alpha=lower_limit, beta=upper_limit, 
                        norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)

# Display the original and adjusted images
cv2_imshow(img)
cv2_imshow(img_adj)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

gray_negative = abs(255-gray)
cv2_imshow(gray_negative)
```

```python
img = cv2.imread('img.jpg', 0)  # Load as grayscale image

# Apply histogram equalization
img_eq = cv2.equalizeHist(img)

# Display the original and equalized images
cv2_imshow(img)
cv2_imshow(img_eq)
```

```python
hist = plt.hist(img.ravel(), bins=256, range=(0, 256))

# Display the histogram
plt.title('Original Image Histogram')
plt.xlabel('Pixel Values')
plt.ylabel('Frequency')
plt.show()

hist = plt.hist(img_eq.ravel(), bins=256, range=(0, 256))

# Display the histogram
plt.title('Eq. Image Histogram')
plt.xlabel('Pixel Values')
plt.ylabel('Frequency')
plt.show()
```

Linear Stretching

```python
contrast = io.imread("wom.jpg", as_gray=True)
plt.imshow(contrast, cmap='gray', norm=NoNorm())
```

```python
max = contrast.max()
min = contrast.min()
print(max, min)
```

```python
ratio = ((255 - 0)/(int(max)-int(min))) + 0
print(ratio)
```

```python
linearContrast = (contrast-int(min))*ratio
plt.imshow(linearContrast, cmap="gray", norm=NoNorm())
```
